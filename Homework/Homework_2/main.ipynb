```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.array([i*np.pi/30.0 for i in range(31)])
y = np.cos(t)

S = 0
for n in range(len(t)):
    S += t[n]*y[n]

print("The sum is: " + str(S))
```

```python
theta = np.linspace(0,2*np.pi,1000)
R = 1.2
r = 0.1
f = 15
p = 0

x = lambda t: R*(1 + r*np.sin(f*t + p))*np.cos(t)

y = lambda t: R*(1 + r*np.sin(f*t + p))*np.sin(t)

plt.plot(x(theta), y(theta))
plt.xlabel("x")
plt.ylabel("y")
plt.title("Wavy Circle")
plt.axis("square")

plt.savefig("Second_Plot")
plt.savefig("First_Plot.png")
```
