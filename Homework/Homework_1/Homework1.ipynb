```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def p1(x):
    return x**9 - 18*x**8 + 144*x**7 - 672*x**6 + 2016*x**5 - 4032*x**4 + 5376*x**3 - 4608*x**2 + 2304*x - 512

def p2(x):
    return (x-2)**9
```

```python
x = np.arange(1.92,2.08,0.001)
y1 = p1(x)
y2 = p2(x)
```

```python
#plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
deltas = np.array([10**(-(16-i)) for i in range(17)])
y_pi = np.cos(np.pi + deltas) - np.cos(np.pi)
y_10 = np.cos(10**6 + deltas) - np.cos(10**6)
p_pi = -2.0*np.sin((2.0*np.pi + deltas)/2.0)*np.sin(deltas/2.0)
p_10 = -2.0*np.sin((2.0*10**6 + deltas)/2.0)*np.sin(deltas/2.0)

m_pi = np.abs(y_pi-p_pi)
m_10 = np.abs(y_10-p_10)

plt.plot(deltas, m_pi, label = "x = pi")
plt.plot(deltas, m_10, label = "x = 10^6")
plt.yscale('log')
plt.xscale('log')
plt.legend()
plt.xlabel("Delta Value")
plt.ylabel("Magnitude of Difference")
plt.title("Original vs Adjusted Magnitudes")
plt.show()
```

```python
n_pi = -deltas*np.sin(np.pi)
n_10 = -deltas*np.sin(10**6)

plt.plot(deltas, np.abs(n_pi-y_pi), label = "x = pi")
plt.plot(deltas, np.abs(n_10-y_10), label = "x = 10^6")
plt.yscale('log')
plt.xscale('log')
plt.legend()
plt.xlabel("Delta Value")
plt.ylabel("Magnitude of Difference")
plt.title("Original vs Taylor Magnitudes")
plt.show()
```
